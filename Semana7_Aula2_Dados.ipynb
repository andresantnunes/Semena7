```python
import pandas as pd

tempo = pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'])
vendas = [100, 150, 130, 210]
categoria = ['Eletrônicos', 'Eletrônicos', 'Roupas', 'Roupas']

df_vendas = pd.DataFrame({
    'tempo': tempo,
    'vendas': vendas,
    'categoria': categoria
})

display(df_vendas.head())
```

```python
import seaborn as sns


sns.set_theme(style="darkgrid")
sns.lineplot(
    x=df_vendas['tempo'],
    y=df_vendas['vendas'],
    hue=df_vendas['categoria'], # hue -> tonalidade
    palette="dark" # Define uma paleta de cores escuras
    )
```

Analise de venda por categoria:
- Objetivo: Entender qual das categorias gerou maior volume de vendas
- Método: Comparação com Gráfico em Barra de Cada Categoria

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.DataFrame({'Produto': ['Alimento', 'Limpeza', 'Bebidas'], 'Qtd': [45, 88, 32]})
sns.barplot(data=dados, x='Produto', y='Qtd', hue='Produto', legend=False)
plt.title('Vendas por Categoria')
plt.show()
```

Podemos ver que os produtos de limpeza geram um volume muito maior de vendas em quantidade

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Gerando dados de exemplo com mais valores
investimento = [100, 300 , 200, 150, 300, 250, 180, 220, 280, 120, 350, 210, 160, 270, 190, 310]
retorno = [10, 10, 25, 18, 40, 30, 20, 28, 35, 15, 45, 22, 19, 38, 24, 42]
alcance = [500, 750, 1200, 800, 1800, 1000, 900, 1300, 1600, 700, 2000, 1100, 850, 1700, 950, 1900]

df = pd.DataFrame({
    'investimento': investimento,
    'retorno': retorno,
    'alcance': alcance
})

sns.scatterplot(x='investimento', y='retorno', data=df, size='alcance', hue='alcance', sizes=(50, 500))
plt.title('Investimento vs Retorno com Alcance')
plt.xlabel('Investimento')
plt.ylabel('Retorno')
plt.show()
```
