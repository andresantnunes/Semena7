```python
import pandas as pd
import seaborn as sns

tempo = pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'])
vendas = [100, 150, 130, 210]
categoria = ['Eletrônicos', 'Eletrônicos', 'Roupas', 'Roupas']

df_vendas = pd.DataFrame({
    'tempo': tempo,
    'vendas': vendas,
    'categoria': categoria
})

print(df_vendas.head())



sns.set_theme(style="darkgrid")
sns.lineplot(
    x=df_vendas['tempo'], 
    y=df_vendas['vendas'], 
    hue=df_vendas['categoria'], # hue -> tonalidade
    palette="magma" # Define uma paleta de cores escuras
    )
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import mplcursors # Biblioteca para adicionar interatividade ao Matplotlib

investimento = [100, 300 , 200, 150, 300, 250, 180, 220, 280, 120, 350, 210, 160, 270, 190, 310]
retorno = [10, 10, 25, 18, 40, 30, 20, 28, 35, 15, 45, 22, 19, 38, 24, 42]
alcance = [500, 750, 1200, 800, 1800, 1000, 900, 1300, 1600, 700, 2000, 1100, 850, 1700, 950, 1900]

df = pd.DataFrame({
    'investimento': investimento,
    'retorno': retorno,
    'alcance': alcance
})

# Salvamos o eixo do gráfico em uma variável 'ax'
ax = sns.scatterplot(x='investimento', y='retorno', data=df, size='alcance', hue='alcance', sizes=(50, 500))
plt.title('Investimento vs Retorno com Alcance')
plt.xlabel('Investimento')
plt.ylabel('Retorno')

# Adicionando o cursor interativo (mostra dados ao passar o mouse ou clicar)
cursor = mplcursors.cursor(ax, hover=True)

# Formatando o texto que aparece na caixa interativa
@cursor.connect("add")
def on_add(sel):
    index = sel.index
    sel.annotation.set_text(
        f"Invest.: {df['investimento'].iloc[index]}\n"
        f"Retorno: {df['retorno'].iloc[index]}\n"
        f"Alcance: {df['alcance'].iloc[index]}"
    )
    sel.annotation.get_bbox_patch().set(alpha=0.8) # Deixa a caixinha um pouco transparente

plt.show()
```
